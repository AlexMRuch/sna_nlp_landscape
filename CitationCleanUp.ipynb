```python
import numpy as np
import pandas as pd 
```

```python
# converting to pandas dataframe 
citations = pd.read_csv('citations.csv')
keywords = pd.read_csv('keywords.csv')
```

## Forming tuples from keywords 

```python
keywords.head(5)
```

```python
keyauthor1 = keywords[['Source Paper', 'Authors']].copy()
l = len(keyauthor1)

def remove(string):
    return string.replace(" ", "")

data = [] 
for x in range(l):
    paper = keyauthor1.loc[x, 'Source Paper']
    authors = keyauthor1.loc[x, 'Authors']
    authors = remove(authors)
    authorslist = authors.split(",")
    for y in range(len(authorslist)):
        data.append({'Source Paper': paper, 'Relation':"is written by", 'Author': authorslist[y]})
        
keypaperauthor = pd.DataFrame(data)
keypaperauthor

keyauthorpaper = keypaperauthor[['Author', 'Source Paper']].copy()
l = len(keyauthorpaper)
r = ["wrote" for _ in range(l)]
keyauthorpaper.insert(1, 'Relation',r)
keyauthorpaper.head(5)
```

```python
df = pd.concat([df, keypaperauthor.rename(columns = {"Source Paper": "Source", "Author": "Destination"})], sort = False, ignore_index = True)
df
```

```python
df1 = pd.concat([df, keyauthorpaper.rename(columns = {"Author": "Source", "Source Paper": "Destination"})], sort = False, ignore_index = True)
df1
```

```python
paperjournal = keywords[['Source Paper', 'Journal']].copy()
journalpaper = keywords[['Journal', 'Source Paper']].copy()
l = len(paperjournal)
r1 = ["was published in the journal" for _ in range(l)]
r2 = ["published the paper" for _ in range(l)]
paperjournal.insert(1, 'Relation',r1)
journalpaper.insert(1, 'Relation',r2)
paperjournal.head(5)
```

```python
df = pd.concat([df1, paperjournal.rename(columns = {"Source Paper": "Source", "Journal" : "Destination"}), journalpaper.rename(columns = {"Journal": "Source", "Source Paper": "Destination"})], sort = False, ignore_index = True)
df 
```

```python
paperkeywords1 = keywords[['Source Paper','Keywords']].copy()

paperkeywords2 = paperkeywords1.replace(np.nan, '', regex=True)

l = len(paperkeywords1)

def splitKeywords(string): 
    s = string.split(",")
    if s == string: 
        return string.split(";")
    else: 
        return s 
def deleteWhitespace(List): 
    l = len(List)
    for x in range(l): 
        List[x] = List[x].strip()
    return List

data = [] 

for x in range(l): 
    paper = paperkeywords2.loc[x, 'Source Paper']
    keyWords = paperkeywords2.loc[x, 'Keywords']
    if (keyWords != ''):         
        keywordslower = keyWords.lower()
        keywordslist = splitKeywords(keywordslower)
        keywordslist = deleteWhitespace(keywordslist)
        for y in range(len(keywordslist)):
            data.append({'Source Paper': paper, 'Relation': "has the keyword", 'Keywords': keywordslist[y]})
    
paperkeywords = pd.DataFrame(data)

keywordspaper = paperkeywords[['Keywords', 'Source Paper']].copy()
l = len(keywordspaper)
r = ["is a keyword in" for _ in range(l)]
keywordspaper.insert(1, 'Relation',r)    
```

```python
df = pd.concat([df, paperkeywords.rename(columns = {"Source Paper": "Source", "Keywords": "Destination"})], sort = False, ignore_index = True)
df
```

```python
df1 = pd.concat([df, keywordspaper.rename(columns = {"Keywords": "Source", "Source Paper": "Destination"})], sort = False, ignore_index = True)
df1
```

## Creating Encoding Identity Maps 

```python
# want to create create two encoding-identity maps: one that maps an encoding number an entity name (e.g., that maps 0 to an
# author/paper/journal/etc. entity) and one that maps an encoding number to a relation name (e.g., that maps 0 to a cites/publ
# ished-in-year/published-in-journal/etc. relation). After making these two encoding-identity maps, we’ll need to apply them to
# the concatenated dataframe (which is now called the canonical tuples dataframe, where the three columns are source, relation, 
# and destination). You’ll save all three of these as a .tsv file.

# Task 1: concatenate the dataframes (saved as df1)
# Task 2: make an entities.tsv 
# Task 3: make an relations.tsv 

```

```python
# "was published in the year" for _ in range(l) r2 = ["was the year of publication of"
r = {'Relations': ["cites", "is cited by", "was published in the year", "was the year of publication of", "wrote", "is written by", "was published in the journal", "published the paper", "has the keyword", "is a keyword in"]}
relations = pd.DataFrame(data = r)
relations.to_csv("relations.tsv", sep="\t")
```

```python
def encoding_map (en, re, df): 
    data = []
    l = df.size 
    for i in range(l):
        source = df.loc[i, 'Source']
        s_code = en[en['Entities'] == source].index.values
        s_code = s_code[0]
        relat = df.loc[i, 'Relation']
        r_code = re[re['Relations'] == relat].index.values
        r_code = r_code[0]
        dest = df.loc[i, 'Destination']
        d_code = en[en['Entities'] == dest].index.values
        d_code = d_code[0]
        data.append({'Source': s_code, 'Relation': r_code, 'Destination': d_code})
    tuples = pd.DataFrame(data)
    return tuples
    
```
